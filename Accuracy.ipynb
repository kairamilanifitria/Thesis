```python
import matplotlib.pyplot as plt

# Data
categories = ['RAG', 'Others']
scores = [96, 55]

# Plotting the bar chart
bars = plt.bar(categories, scores, color=['#FF9999', '#66B2FF'])

# Adding value annotations above each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval}%', ha='center', va='bottom', color='black')

# Adding a title and labels
plt.title('Score Comparison between RAG and Others (Our Datasets)')
plt.xlabel('Categories')
plt.ylabel('Scores')


# Adjusting the layout to provide space for the title
plt.tight_layout()

# Display the chart
plt.show()
```
